```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import smtplib
from email.message import EmailMessage

Sender_email = 'abcd1234@gmail.com'
Password = 'Abcd123'
Reciever_email = 'pydev@gmailcom'

data = pd.read_csv("sales_data_sample.csv", encoding='latin1')
#print("Missing values before Drop & Fill\n" , data.isnull().sum())

data.drop(columns = ['ADDRESSLINE2','STATE','POSTALCODE'], inplace = True , errors='ignore')
data['TERRITORY'].fillna("Unknown Territory" , inplace = True)

#print("Missing values After Drop & Fill\n" , data.isnull().sum())

data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
data['Month'] = data['ORDERDATE'].dt.month

monthly_sales = data.groupby(['Month', 'PRODUCTLINE'])['SALES'].sum().unstack()
top_products = data.groupby('PRODUCTLINE')['SALES'].sum().nlargest(5).reset_index()
region_sales = data.groupby('TERRITORY')['SALES'].sum().reset_index()

monthly_sales.plot(kind='line', marker='o', figsize=(10,6))
plt.title("Monthly Sales by Product Line")
plt.ylabel("Sales ($)")
plt.xlabel("Month")
plt.grid(True)
plt.tight_layout()
plt.savefig("sales_trend_chart.png")
plt.close()

top_products.plot(kind='bar', color='orange')
plt.title("Top 5 Product Lines by Total Sales")
plt.ylabel("Sales ($)")
plt.tight_layout()
plt.savefig("top_products_chart.png")
plt.close()

file = "Sales Report.xlsx"

if os.path.exists(file):
    os.remove(file)

with pd.ExcelWriter(file, engine = 'xlsxwriter') as writer:
    monthly_sales.to_excel(writer, sheet_name ='Monthly Sales by Product')
    top_products.to_excel(writer, sheet_name = 'Top 5 Products', index = False)
    region_sales.to_excel(writer, sheet_name = 'Region Sales', index = False)

    workbook = writer.book
    worksheet1 = writer.sheets['Monthly Sales by Product']
    worksheet2 = writer.sheets['Top 5 Products']
    worksheet1.insert_image('J2', 'sales_trend_chart.png')
    worksheet2.insert_image('F2', 'top_products_chart.png')

attachments = ['Sales report.xlsx', 'sales_trend_chart.png', 'top_products_chart']

msg = EmailMessage()
msg['Subject']= 'Automated Sales Report'
msg['From']= Sender_email
msg['To']= Recievers_email
msg.set_content ('Dear Sir/Madam,\n\nThis is an automatically generated sales report.Please find the attached file and charts. Regards,\n\nAotomate Bot')

with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
    smtp_login(Sender_email,password)
    smtp.send_message(msg)

print('Email Sent Successfully.')
```
